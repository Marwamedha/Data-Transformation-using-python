Handling Missing Data - fillna, interpolate, dropna

```python
new_df =df2.fillna(0)
new_df
```

```python
new_df = df2.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
new_df = df2.fillna(method="ffill") # RowsMethod of Handling Missing Data 
new_df 
```

```python
new_df =df2.fillna(method='bfill') # Rows Method 
new_df
```

```python
new_df= df2.fillna(method ='ffill',limit =1)
new_df
```

```python
new_df=df2.interpolate() #Interpolation Mathematical Method 
new_df
```

```python
new_df =df2.interpolate('time')
new_df
```

```python
new_df=df2.dropna()
new_df
```

```python
new_df = df2.dropna(how='all') # all data is missing  in the Row 
new_df
```

```python
new_df = df2.dropna(thresh=1)
new_df
```
